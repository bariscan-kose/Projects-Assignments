```python
import requests
import pandas as pd
from io import StringIO
from datetime import datetime
```

```python

EODHD_API_KEY=#REDACTED
curr_pair="MSFT"
exchange="US"

def get_eod_data(symbol=f"{curr_pair}.{exchange}", api_token=EODHD_API_KEY, session=None):
	if session is None:
		session = requests.Session()
		url = f"https://eodhistoricaldata.com/api/eod/{symbol}"
		params = {"api_token": api_token}
		r = session.get(url, params=params)

		if r.status_code == requests.codes.ok:
			df = pd.read_csv(StringIO(r.text), skipfooter=0, parse_dates=[0],engine="python")
		else:
			raise Exception(r.status_code, r.reason, url)
		return df

response_df=get_eod_data()
response_df=response_df.iloc[::-1].reset_index(drop=True)
print("Data length : ", len(response_df))
response_df.iloc[:10]
```

```python
response_df["AdjustedClose"]=response_df["Adjusted_close"]
response_df.drop("Adjusted_close", axis=1, inplace=True)
response_df=response_df.loc[response_df["Date"].dt.dayofweek <5].reset_index(drop=True)
# response_df=response_df.iloc[:3360]
response_df.iloc[:10]
```
