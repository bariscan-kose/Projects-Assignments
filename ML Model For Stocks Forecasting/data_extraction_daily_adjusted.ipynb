```python
#DAILY ADJUSTED

url = "https://alpha-vantage.p.rapidapi.com/query"
symbol="MSFT"
querystring = {"function":"TIME_SERIES_DAILY_ADJUSTED","symbol": symbol ,"outputsize":"full","datatype":"json"}

headers = {
	"X-RapidAPI-Key": "0b49086ad7mshd8eb6d78dfa9981p14f385jsne4bd6ac71584",
	"X-RapidAPI-Host": "alpha-vantage.p.rapidapi.com"
}

response = requests.request("GET", url, headers=headers, params=querystring)
response_json=response.json()

```

```python
response_json
```

```python
response_ts=response_json["Time Series (Daily)"]
response_ts_2=[]
for k,v in response_ts.items():
    response_ts_2.append([k,v["1. open"],v["2. high"],v["3. low"],v["4. close"],
						  v["5. adjusted close"], v["6. volume"],v["7. dividend amount"],
						  v["8. split coefficient"]
						  ])
response_df=pd.DataFrame(response_ts_2, columns=["Index","Open","High","Low","Close","AdjustedClose", "Volume", "DividendAmount", "SplitCoefficient"])
response_df.to_csv("HistoricalData/"+response_json["Meta Data"]["2. Symbol"]+"_DailyAdjusted.csv", index=False)
# response_ts_2
```
